```python
import pdfplumber
import pandas as pd
```

```python
pdf_path="CoffeeShopPriceList.pdf"
all_items=[]

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        table=page.extract_table()
        if table:
            for row in table:
              row= [cell.strip() if cell else "" for cell in row]

              if row[0].lower()=="description" and row[1].lower() == "price":
                if len(row) >= 4 and row[2].lower()== "description" and row[3].lower() == "price": continue

                if len(row) >= 2 and row[0] and row[1]:
                  all_items.append({"description": row[0], "price": row[1]})

                  if len(row) >= 4 and row[2] and row[3]:
                    all_items.append({"description": row[2], "price": row[3]})

```

```python
df=pd.DataFrame(all_items)
df.to_csv("coffee_shop_price.csv", index=False)
```

```python
df=pd.read_csv("coffee_shop_price.csv")
```

```python
df.columns = df.columns.str.strip().str.lower()
df["price"] = df["price"].str.replace(r"\$", "", regex=True).astype(float)
```
